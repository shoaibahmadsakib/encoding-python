# ordinal

```python
df6.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
sex = df6.sex.unique()
smoker = df6.smoker.unique()
region = df6.region.unique()
order = OrdinalEncoder(categories=[sex,smoker,region])
encoded = order.fit_transform(df6[['sex','smoker','region']])
encoded
```

```python
df6[['sex','smoker','region']]=encoded
```

```python
df6.head(10)
```

```python
for col in df6.columns:
        if is_numeric_dtype(df6[col]):
            continue
        else:
            col_catagorize=df6[col].unique()
            order = OrdinalEncoder(categories=[col_catagorize])
            encoded = order.fit_transform(df6[col])
            df6[col]=encoded
```
